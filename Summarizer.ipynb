```python
from openai import OpenAI
from IPython.display import Markdown, display
from scrapper import get_page_content
```

```python
system_prompt="""
You are a snarky assistant that analyzes the contents of a website,
and provides a short, snarky, humorous summary, ignoring text that might be navigation related.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""
```

```python
user_prompt_pre=f"""Here are the contents of a website.
Provide a short summary of this website.
If it includes news or announcements, then summarize these too.

"""
```

```python
def message_for_LLM(data):
    messages=[{"role":"system","content":system_prompt},
              {"role":"user","content":user_prompt_pre+data}]
    return messages
```

```python
def call_model(url):
    data=get_page_content(url)
    MODEL='llama3.2'
    openai=OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
    response = openai.chat.completions.create(
    model=MODEL,
    messages=message_for_LLM(data)
        )
    return response.choices[0].message.content
```

```python

def display_markdown(url):
    summary=call_model(url)
    display(Markdown(summary))
```

```python
url=input("Enter URL to summarize: ")
display_markdown(url)
```
